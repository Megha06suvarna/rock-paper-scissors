```python
import random
choices=["rock","paper","scissors"]

   
user=input("enter the rock,paper,scissors:").lower()
if user=="quit":
    break
    
computer=random.choice(choices)

print("compurter choice:",computer)

if user==computer:
    print("its a tie!")
elif user=="rock":
    if computer=="scissors":
        print("you win!")
    else:
        print("you lose!")
        
elif user=="paper":
    if computer=="rock":
        print("you win!")
    else:
        print("you lose!")
        
elif user=="scissors":
    if computer=="paper":
        print("you win!")
    else:
        print("you lose!")
        
else:
    print("invalid input")
        
```
