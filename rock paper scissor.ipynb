```python
import random

choices = ["rock", "paper", "scissors"]

while True:
    user = input("Enter rock, paper, scissors or quit: ").lower()

    if user == "quit":
        print("Game ended")
        break

    computer = random.choice(choices)
    print("Computer choice:", computer)

    if user == computer:
        print("It's a tie!")
    elif user == "rock":
        if computer == "scissors":
            print("You win!")
        else:
            print("You lose!")
    elif user == "paper":
        if computer == "rock":
            print("You win!")
        else:
            print("You lose!")
    elif user == "scissors":
        if computer == "paper":
            print("You win!")
        else:
            print("You lose!")
    else:
        print("Invalid input")
```
